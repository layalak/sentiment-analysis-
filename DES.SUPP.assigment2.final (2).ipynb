```python
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import os
from itertools import count
from nltk.util import pr
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import csv
sentiments = SentimentIntensityAnalyzer()
reviewsstc['positive'] = [sentiments.polarity_scores(i)["pos"] for i in reviewsstc["content"]]
reviewsstc['negative'] = [sentiments.polarity_scores(i)["neg"] for i in reviewsstc["content"]]
reviewsstc['neutral'] =  [sentiments.polarity_scores(i)["neu"] for i in reviewsstc["content"]]

#reviewsstc['negative'] = round([sentiments.polarity_scores(i)["neg"]* 10,2) for i in reviewsstc["content"])
#reviewsstc['neutral'] = round([sentiments.polarity_scores(i)["neu"]* 10,2) for i in reviewsstc["content"])
```

```python
!pip install TextBlob
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
reviewsstc['score'] = reviewsstc['content'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
reviewsstc.head()
```

```python
reviewsstc['sentiment'] = reviewsstc['score'].apply(lambda x: 'positive' if x > 0.0 else 'neutral' if x == 0.0 else 'negative' )
```

```python
reviewsstc.head()
```

```python
value = reviewsstc['sentiment'].value_counts().values
index = reviewsstc['sentiment'].value_counts().index
```

```python
plt.title('sentoemt analysis using')
plt.xlabel('sentiment')
plt.ylabel('number of sentiemnt')
plt.bar(index, value, color=['#32a852','#325ba8','#a83289'], alpha=0.7)
plt.grid(color='#a83289', linestyle='--',linewidth=1,axis='y',alpha=0.25)
plt.show
```
